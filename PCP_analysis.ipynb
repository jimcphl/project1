```python
import os
import pandas as pd

#url = "S:\ENROLLMENT\ENROLLMENT-PUBLIC\MEDICARE\MAP\Source\MAP Roster_2021- KAT2.xlsx"
url = "C:\\Users\\tflung\\Downloads\\PCP_analysis.xlsx"

df1=pd.read_excel(
     os.path.join(url),
     engine='openpyxl',
     sheet_name='Not MGNA_ PCP'
)

df2=pd.read_excel(
     os.path.join(url),
     engine='openpyxl',
     sheet_name='Not MGNA_ Specialist'
)

print (df2.columns)
```

```python
df3 = df1[['NPI', 'PROVIDER FULL NAME', 'County', 'STREET ADDRESS', 
           'City', 'State', 'Zip code', 'Accepting New Patients', 
           'category', 'Specialty', 'Site Name', 'Gender', 'Language']]

df3 = df3[df3['State'] == 'NY']

df4 = df2[['NPI', 'PROVIDER FULL NAME', 'County', 'STREET ADDRESS', 
           'City', 'State', 'Zip code', 'Accepting New Patients', 
           'category', 'Specialty', 'Site Name', 'Gender', 'Language']]

df4 = df4[df4['State'] == 'NY']

#print (df4['County'].unique())

df4 = df4[(df4['County'] == 'New York')
          | (df4['County'] == 'Queens')
          | (df4['County'] == 'Kings')
          | (df4['County'] == 'Bronx')
          | (df4['County'] == 'Nassau')
          | (df4['County'] == 'Westchester')
         ]

print (len(df4))
```

```python
df3.to_csv(r'PCP1.csv', index = False)

df4.to_csv(r'PCP2.csv', index = False)
```
