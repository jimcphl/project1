```python
import pandas as pd
import os

# DualEligibleMembers.xlsx
#pd.read_excel('DualEligibleMembers.xlsx', index_col=0)

url = "C:\\Users\\tflung\\Downloads\\EventsMaster.xlsx"

#url = "MAPD&SNP Roster_2021.xlsx"

df1=pd.read_excel(
     os.path.join(url),
     engine='openpyxl',
)
df1
```

```python
# Rename

df2 = df1.rename(columns={'Primary Column': 'Primary'})
df2 = df2.rename(columns={'Event Date': 'Date'})
df2 = df2.rename(columns={'Campaign Initiative': 'Campaign'})
df2 = df2.rename(columns={'Event Name': 'Name'})
df2 = df2.rename(columns={'Event Site Type': 'Site'})
df2 = df2.rename(columns={'Site Contact Name': 'Contact'})
df2 = df2.rename(columns={'Agent(s) Assigned': 'Agent'})
df2 = df2.rename(columns={'Logistics Required': 'Logistics'})
df2 = df2.rename(columns={'SET UP TIME': 'SETUPTIME'})
df2 = df2.rename(columns={'# OF LEADS/PTC COLLECTED': 'LEADS'})
df2 = df2.rename(columns={'NUMBER OF SALES/ENROLLMENTS': 'ENROLLMENTS'})
df2 = df2.rename(columns={'NUMBER OF APPOINTMENTS SCHEDULED IF ANY': 'APPOINTMENTS'})
df2 = df2.rename(columns={'ROI (NUMBER OF SALES / NUMBER COST)': 'COST_ROI'})
df2 = df2.rename(columns={'OPPORTUNITY TYPE': 'OPPORTUNITY'})
df2 = df2.rename(columns={'EVENTS COMMENTS/RATIONAL': 'COMMENTS'})
df2 = df2.rename(columns={'Cost': 'COST'})

# select columns

df3 = df2[['Date', 'Campaign', 'Name', 'Site', 'ZIP',
           'County', 'Contact', 'Agent', 'Logistics', 
           'SETUPTIME', 'LEADS', 'ENROLLMENTS', 'APPOINTMENTS',
           'COST', 'OPPORTUNITY']]

df3
```

```python
df3 = df3[df3.Agent != 'Hicham']

df3 = df3.dropna(subset=['LEADS'])

#df3 = df3.dropna(subset=['LEADS', 'ENROLLMENTS', 'APPOINTMENTS'])

#c = df3.loc['LEADS', 'ENROLLMENTS','APPOINTMENTS'].columns  # retrieve only the 0th row for efficiency
#df3 = df3.dropna(subset=c, how='all')



df3

```
