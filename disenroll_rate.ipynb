```python
# df28, df30, df27, df16

# change type of month as date
df16['Month']= pd.to_datetime(df16['Month'])
df27['Month']= pd.to_datetime(df27['Month'])

df25 = df16[['Month', 'D-SNP', 'MAPD', 'MAP']]

# change type of number
df25['D-SNP'] = df25['D-SNP'].astype(str).astype(int)
df25['MAPD'] = df25['MAPD'].astype(str).astype(int)
df25['MAP'] = df25['MAP'].astype(int)

# add conditional value
df25['CENSUS'] = df25['D-SNP'] + df25['MAPD'] + df25['MAP']

# merge 2 dataframe
df31 = df25.merge(df27, how='left', on='Month')
df31['Overall_DISEN'] = df31['D-SNP_Loss'] + df31['MAPD_Loss'] + df31['MAP_Loss']

# Change data
df31['Month'] = df31['Month'] + pd.to_timedelta(1, unit='d')

df31 = df31.merge(df28, how='left', on='Month')
df31 = df31.merge(df30, how='left', on='Month')
df31 = df31[['Month','CENSUS', 'Overall_DISEN', 'EN', 'RDE']]



df31['RAPID_DIS_RATE'] = (df31['RDE'] / df31['EN']).round(2)

df31['OVERALL_DIS_RATE'] = (df31['Overall_DISEN'] / df31['CENSUS']).round(2)

df31['RAPIDDIS_OVERALLDIS'] = (df31['RDE'] / df31['Overall_DISEN']).round(2)

df31 = df31.fillna(0)

df31

```
