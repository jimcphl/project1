```python
import pandas as pd
import os

# DualEligibleMembers.xlsx
#pd.read_excel('DualEligibleMembers.xlsx', index_col=0)

url = 'C:/Users/tflung/Downloads/DisenrollmentSurvey.xlsx'

#url = "MAPD&SNP Roster_2021.xlsx"

df1=pd.read_excel(
     os.path.join(url),
     engine='openpyxl',
)
#df1

#print(list(df1.columns))

# Rename

df2 = df1.rename(columns={'Service Area': 'Area'})
df2 = df2.rename(columns={'Enrollment Status': 'Status'})
df2 = df2.rename(columns={'Enrollment Date': 'EnrollDate'})
df2 = df2.rename(columns={'Disenrollment/ End Date': 'DisenrollDate'})
df2 = df2.rename(columns={'App Date': 'AppDate'})
df2 = df2.rename(columns={'Sales Team': 'Team'})
df2 = df2.rename(columns={'Licensed Benefits Agent': 'Agent'})
df2 = df2.rename(columns={'Outreach Associate\xa0Conducting Call': 'OutreachCall'})
df2 = df2.rename(columns={'Q1= Disenrollment Reason Code': 'Q1'})
df2 = df2.rename(columns={'Q2= PCP NETWORK MEMBER RATING': 'Q2'})
df2 = df2.rename(columns={'Q3= SPECIALIST NETWORK MEMBER RATING': 'Q3'})
df2 = df2.rename(columns={'Q4= CUSTOMER SERVICE DEPARTMENT QUALITY RATING': 'Q4'})
df2 = df2.rename(columns={'Q5= BENEFIT ADVISOR RATING': 'Q5'})

# select columns

df3 = df2[['EnrollDate', 'Language', 'Area', 'LOB', 'Status', 'DisenrollDate', 'AppDate', 'Team', 'Agent', 'OutreachCall', 'Q1', 'Q2', 'Q3', 'Q4', 'Q5']]

df3 = df3.fillna('blank')
print(list(df3.columns))
```
