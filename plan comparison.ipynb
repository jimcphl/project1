```python
import bs4 as bs
import urllib.request
import pandas as pd
from tabulate import tabulate


# https://q1medicare.com/MedicareAdvantage-PartC-MedicareHealthPlanBenefits.php?state=NY&source=2016MAFinder&ZIP=&countyCode=36085&contractId=H3359&planId=001&segmentId=0&plan=Healthfirst%2065%20Plus%20Plan%20(HMO)%20-%20H3359-001-0&utm_source=partd&utm_medium=mafinder&utm_campaign=enrollbutton


source = urllib.request.urlopen('https://q1medicare.com/MedicareAdvantage-2020C-MedicareHealthPlanBenefits.php?state=NY&source=2020MAFinder&ZIP=&countyCode=36029&contractId=H3387&planId=010&segmentId=0&plan=UnitedHealthcare%20Dual%20Complete%20(HMO%20D-SNP)%20-%20H3387-010-0&utm_source=partd&utm_medium=mafinder&utm_campaign=planname').read()


#  source = urllib.request.urlopen('').read()
#  source = urllib.request.urlopen('').read()


#  source = urllib.request.urlopen('').read()
#  source = urllib.request.urlopen('').read()


#  source = urllib.request.urlopen('').read()
#  source = urllib.request.urlopen('').read()

soup = bs.BeautifulSoup(source,'lxml')

table = soup.find('div', attrs={'class':'resultchart'})
table_rows = table.find_all('tr')

res = []
for tr in table_rows:
    td = tr.find_all('td')
    row = [tr.text.strip() for tr in td if tr.text.strip()]
    if row:
        res.append(row)

res
```
