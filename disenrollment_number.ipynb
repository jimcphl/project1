```python
# change type of month as date
df16['Month']= pd.to_datetime(df16['Month'])

df25 = df16[['Month', 'D-SNP', 'MAPD', 'MAP']]

# change type of number
df25['D-SNP'] = df25['D-SNP'].astype(str).astype(int)
df25['MAPD'] = df25['MAPD'].astype(str).astype(int)
df25['MAP'] = df25['MAP'].astype(int)

# add conditional value
df25['Total_Gain'] = df25['D-SNP'] + df25['MAPD'] + df25['MAP']

# merge 2 dataframe
df28 = df25.merge(df27, how='left', on='Month')
df28['Actual_loss'] = df28['D-SNP_Loss'] + df28['MAPD_Loss'] + df28['MAP_Loss']

# balance
df28['Cum_Balance'] = df28['Total_Gain'].cumsum().shift() - df28['Actual_loss'].cumsum().shift()
#df28['Balance'] = df28['Total_Gain'].shift() - df28['Actual_loss'].shift()
#df28['Cum_Balance'] = df28['Cum_Balance'].astype(int)
df28['Cum_Balance'] = df28['Cum_Balance'].fillna(0)

# Change data
df28['Month'] = df28['Month'] + pd.to_timedelta(1, unit='d')

df35 = df34[['Month', 'D-SNP_teamGain', 'MAPD_teamGain']]

df28 = df28.merge(df35, how='left', on='Month')

df37 = df24
df37['Month'] = df37['Month'] + pd.to_timedelta(1, unit='d')
df37 = df37.rename(columns={'Map_Gain': 'MAP_teamGain'})


df28 = df28.merge(df37, how='left', on='Month')
df28['TeamGain'] = df28['D-SNP_teamGain'] + df28['MAPD_teamGain'] + df28['MAP_teamGain']

# (Field 8 agents, 3 Telesales, 10 FMO broker, WEB)
df28.loc[df28['Month'] == '2021-01-01', 'EnrollTarget'] = (45*8) + (45*3) + 100 + 75
df28.loc[df28['Month'] != '2021-01-01', 'EnrollTarget'] = 18 + (13*7) + (13*3) + 20 + 25

df28.loc[df28['Month'] == '2021-01-01', 'Target_Disenroll%'] = 0.05
df28.loc[df28['Month'] != '2021-01-01', 'Target_Disenroll%'] = 0.02

df28.loc[df28['Month'] == '2021-01-01', 'DisenrollTarget'] = ((df28['D-SNP'] + df28['MAPD'] + df28['MAP'])*0.05).round(0)
df28.loc[df28['Month'] != '2021-01-01', 'DisenrollTarget'] = ((df28['D-SNP'] + df28['MAPD'] + df28['MAP'])*0.02).round(0)

#df28.fillna(0)



df28.loc[df28['Month'] == '2021-01-01', 'DisenrollTarget'] = df28['Cum_Balance'] * 0.05
df28.loc[df28['Month'] != '2021-01-01', 'DisenrollTarget'] = df28['Cum_Balance'] * 0.02
   
# change float to int
df28['DisenrollTarget'] = df28['DisenrollTarget'].astype(int)

df28
```
